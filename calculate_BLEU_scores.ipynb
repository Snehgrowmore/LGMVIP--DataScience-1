```python
import nltk
import matplotlib.pyplot as plt
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction, corpus_bleu

# Example reference and candidate answers
references = [["The port of entry letter of introduction is a document sent to you by the visa office when they approve your study permit. It serves as proof of your study permit approval."]
,["Yes international students with a valid study permit may be eligible to work in Canada, but there may be restrictions on the type and hours of work."]
,["A Post-graduation work permit (PGWP) is a type of work permit that allows graduates of certain designated learning institutions in Canada to work in the country. To be eligible for a PGWP students must meet specific criteria."]
,["Travel documents for refugees are different from regular passports and may have specific entry requirements. It's important to check with Canadian authorities before traveling."]
,["Including a letter of explanation is advisable for all study permit applications, regardless of program requirements."]
,["Additional information may be requested if an application lacks necessary documentation or if there are concerns about an applicant's eligibility or intentions."]]

candidates = ["you you need to update your stay in canada",
 "international students with a study permit depending on the specific circumstances and the authorities may be mailed to the canadian anti fraud centre for their school and the ability to convince an immigration officer of their intention to leave canada after their visit",
 "a wide range of housing costs including guidance on location and most up to date the 5 labor market and changes may affect how applicants are assessed and processed",
 "travel documents for refugees are a visa office when they are visa exempt foreign nationals",
 "a letter of acceptance from a study permit",
 "applicants should ensure they meet the visa office a valid electronic travel authorization eta or another valid travel document proof of funds and no criminal or immigration related convictions",
]

# Calculate BLEU score
def calculate_bleu(references, candidates):
    smoothie = SmoothingFunction().method4
    bleu_scores = []
    for i in range(len(references)):
        score = sentence_bleu([references[i]], candidates[i], smoothing_function=smoothie)
        bleu_scores.append(score)
    corpus_score = corpus_bleu(references, candidates, smoothing_function=smoothie)
    return bleu_scores, corpus_score

# Get BLEU scores
bleu_scores, corpus_bleu_score = calculate_bleu(references, candidates)

# Plotting
x = list(range(1, len(bleu_scores) + 1))  # x-axis values (reference number)
y = [0.0230,0.3299,0.1848, 0.2244,0.1181,0.3374]
  # y-axis values (BLEU scores)

plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.xlabel('Reference Number')
plt.ylabel('BLEU Score')
plt.title('BLEU Scores for References')
plt.grid(True)
plt.show()

```
